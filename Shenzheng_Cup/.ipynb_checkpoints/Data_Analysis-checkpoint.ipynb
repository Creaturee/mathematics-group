```python
def buffer_time(data,a,b,c,d):
    rtt=data.iloc[:,2]
    max_speed=data.iloc[:,1]
    return data.iloc[:,13]*(1+a*rtt)*(1+b)/(c*max_speed-d*rtt)
```

```python
def buffer_time_2(data,a,b,c,d):
    rtt=data.iloc[:,2]
    max_speed=data.iloc[:,1]
    return data.iloc[:,13]*(1+a*rtt)*(1+b)/(c*max_speed/(d*rtt))
```

```python
def line_buffer(data,a):
    max_speed=data.iloc[:,1]
    return data.iloc[:,13]/(a*max_speed)
```

```python
def eval_fit(buffer_time,data):
    model_1, pcov=op.curve_fit(buffer_time,data,data.iloc[:,4])
    print('model')
    print(model_1)
    print('std_error')
    print(np.mean(np.sqrt(np.power(data.iloc[:,4]-buffer_time(data,*model_1),2))))
    print('describe')
    print((data.iloc[:,4]-buffer_time(data,*model_1)).describe())
```

```python
eval_fit(buffer_time,data)
```

```python
eval_fit(line_buffer,data)
```

```python
eval_fit(buffer_time_2,data)
```
